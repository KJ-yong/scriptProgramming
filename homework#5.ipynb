```python
class Rectangle :
    def __init__(self, w, h) :
        self.w = w
        self.h = h
    def __str__(self) : 
        return "너비가 %d, 높이가 %d인 직사각형 입니다." %(self.w,self.h)

class Square(Rectangle) :
    def __init__(self, w) :
        Rectangle.__init__(self,w,w)
    def __str__(self) :
        return "한변의 길이가 %d인 정사각형 입니다." %(self.w)
    
r = Rectangle(2,4)
s = Square(5)
print(r)
print(s)
```

### 3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)
#### 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
##### count: Counter 인스턴스가 지니는 초기 정수 값
##### step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)
##### >>> c = Counter(10)
##### >>> d = Counter(10, 2)
#### 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.
##### >>> print(c)
##### [Count (step: 1)] 10
##### >>> print(d)
##### [Count (step: 2)] 10
#### 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.
##### >>> c.incr()
##### >>> d.incr()
##### >>> print(c)
##### [Count (step: 1)] 11
##### >>> print(d)
##### [Count (step: 2)] 12
#### 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.
##### >>> c()
##### >>> d()
##### >>> print(c)
##### [Count (step: 1)] 12
##### >>> print(d)
##### [Count (step: 2)] 14
#### 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
##### >>> c = c + 5
##### >>> d = d - 5
##### >>> print(c)
##### [Count (step: 1)] 17
##### >>> print(d)
##### [Count (step: 2)] 9
#### 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
##### >>> print(c) > 10
##### True
##### >>> print(d) > 10
##### False
##### >>> print(c) < 10
##### False
##### >>> print(d) < 10
##### True
##### >>> print(c) == 17
##### True
##### >>> print(d) != 9
##### False

```python
class Count :
    def __init__(self,count,step = 1) :
        self.count = count
        self.step = step
        
    def __str__(self) :
        return "[Counter (step : %d)] %d" %(self.step, self.count)
    def incr(self) :
        self.count += self.step
    def __call__(self) :
        self.incr()
    def __add__(self, i) :
        self.count += i
        return self
    def __sub__(self, i) :
        self.count -= i
        return self
    def __eq__(self, other) :
        if type(other) == type(self) and self.count == other.count : return True
        elif self.count == other : return True
        else : return False
    def __lt__(self, other) :
        if type(other) == type(self) and self.count < other.count : return True
        elif self.count < other : return True
        else : return False
    def __le__(self, other) :
        if type(other) == type(self) and self.count <= other.count : return True
        elif self.count <= other : return True
        else : return False
    def __gt__(self, other) :
        if type(other) == type(self) and self.count > other.count : return True
        elif self.count > other : return True
        else : return False
    def __ge__(self, other) :
        if type(other) == type(self) and self.count >= other.count : return True
        elif self.count >= other : return True
        else : return False

c = Count(10)
d = Count(10,2)
print(c)
print(d)
c.incr()
d.incr()
print(c)
print(d)
c()
d()
print(c)
print(d)
c = c+5
d = d-5
print(c)
print(d)
print((c>10))
print((d>10))
print((c<10))
print((d<10))
print((c==17))
print((d!=9))
```

### 4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
#### MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

#### eliminate_duplicate는 정적 메소드로 인스턴스를 생성하지 않아도 호출할 수 있는 메소드로 리스트를 인수로 받아 리스트의 원소에 중복된 원소가 있으면 중복인 원소를 제거한다.
#### __init__은 해당 클래스의 인스턴스가 생성될 때 호출되는 메소드로 리스트를 인수로 받아 추가하고, eliminate_duplicate를 호출하여 중복을 제거한다.
#### __str__은 인스턴스를 문자열로 변환하는 함수로 "MySet : { 인스턴스의 원소 }" 의 형태의 문자열로 반환한다.

### 5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.
#### 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
#### 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
##### >>> u = s | t
##### >>> print(u)
##### MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}
#### 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
##### >>> u = s & t
##### >>> print(u)
##### MySet: {2, 3}
#### 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
##### >>> s = MySet([1, 2, 3])
##### >>> t = MySet([3, 4, 5])
##### >>> u = s - t
##### >>> print(u)
##### MySet: {1, 2}

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, other) : # |연산으로 합집합 구현
        result = self + other
        MySet.eliminate_duplicate(result)
        return result
    
    def __and__(self, other) : # &연산으로 교집합 구현
        result = []
        for e in self :
            if e in other : result.append(e)
        return result
    
    def __sub__(self, other) : # - 연산으로 차집합 구현
        result = []
        for e in self :
            if not (e in other) : result.append(e)
        return result
    
s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
u = s+t
print(u)
u = s & t
print(u)
s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
u = s - t
print(u)
```

```python
import requests

class SearchEngine :
    
    def __init__(self, *urls) :
        self.url_list = list(urls)
        self.stopWords = set(SearchEngine.removeHTML('https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt'))
        try :
            with open("stop_words_english.txt","r", encoding = 'UTF8') as f :
                while True :
                    word = f.readline().strip()
                    if not word : break
                    else :
                        self.stopWords.add(word)
        except FileNotFoundError :
            pass
        
    def addUrl(self, url) :
        self.url_list.append(url)
        
    def removeUrl(self, url) :
        if url in self.url_list :
            self.url_list.remove(url)
    def listUrls(self) :
        for url in self.url_list :
            print(url)
            
    def getWordsFrequency(self) :
        frequency_dict = {}
        punctuation = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~" #구두문자
        for url in self.url_list :
            word_list = SearchEngine.removeHTML(url)
            for w in word_list :
                word = ""
                for c in w : #구두문자 제거
                    if not c in punctuation : word += c
                if not word in self.stopWords : #불용어 제거
                    if frequency_dict.get(word) == None :
                        frequency_dict[word] = 1
                    else : 
                        value = frequency_dict.get(word)
                        frequency_dict[word] = value + 1
        del frequency_dict[''] #공백 제거
        return frequency_dict
    
    def getMaxFreqencyWords(self) :
        if not self.url_list : return None
        else :
            frequency_dict = self.getWordsFrequency()
            max_dict = {}
            maxValue = max(frequency_dict.values())
            for key in frequency_dict.keys() :
                if frequency_dict[key] == maxValue :
                    max_dict[key] = frequency_dict[key]
            return max_dict
    
    def searchUrlByWord(self, search) :
        dict_list = []
        similar_list = []
        punctuation = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~" #구두문자
        if not self.url_list : return None
        else :
            for url in self.url_list :
                word_list = SearchEngine.removeHTML(url)
                url_dict = {}
                for w in word_list :
                    word = ""
                    for c in w : #구두문자 제거
                        if not c in punctuation : word += c
                    if not word in self.stopWords : #불용어 제거
                        if url_dict.get(word) == None :
                            url_dict[word] = 1
                        else : 
                            value = url_dict.get(word)
                            url_dict[word] = value + 1
                del url_dict[''] #공백제거
                dict_list.append(url_dict)
            for word_dict in dict_list :
                search_result = word_dict.get(search)
                if search_result == None : search_result = 0
                similar_list.append(search_result/sum(word_dict.values()))
            max_similar = max(similar_list)
            for i,similar in enumerate(similar_list, 0) :
                if similar == max_similar :
                    print(self.url_list[i])
        
    @staticmethod
    def removeHTML(url) : #url을 받아 html요소를 제외한 단어 리스트 반환
        source = requests.get(url).text
        start_index = 0
        a = source.find("<script", start_index)
        while(a != -1) : #javascript 제거
            b = source.find("</script>",a)
            source = source.replace(source[a:b+9], ' ')
            start_index = a
            a = source.find("<script", start_index)
        start_index = 0
        a =source.find("<style",start_index)
        while(a != -1) : #css 제거
            b = source.find("</style>",a)
            source = source.replace(source[a:b+8],' ')
            start_index = a
            a =source.find("<style",start_index)
        start_index = 0
        a = source.find("<", start_index)
        while(a != -1) : #html태그제거
            b = source.find(">",a)
            if b == -1 : break
            source = source.replace(source[a:b+1],' ')
            start_index = a
            a = source.find("<", start_index)
        return source.split()

```

### 3. 다음과 같은 코딩이 가능하도록 SearchEngineWithOrderedWebWords안에 반복자와 관련된 메소드를 추가하시오.
##### >>> for i in w4:
##### >>>       print(i)
##### ('site, 12)
##### ('science', 11)
##### ('hello', 8)
##### ('world', 2)
##### ('program', 1)
##### ('python', 1)

```python
class SearchEngineWithOrderedWebWords(SearchEngine) :
    def __init__(self, *urls) :
        SearchEngine.__init__(self,*urls)
        self.iterCount = 0
    def getWordsFrequency(self,reverse = False) :
        frequency_dict = super().getWordsFrequency()
        frequency_list = list(frequency_dict.items())
        if not reverse : 
            frequency_list.sort(key = lambda l : l[1], reverse = True)
        else :
            frequency_list.sort(key = lambda l : l[1], reverse = False)
        return frequency_list
    
    def __iter__(self) :
        self.iterCount = 0
        return self
    
    def __next__(self) :
        frequency_list = self.getWordsFrequency()
        if self.iterCount == len(frequency_list) : raise StopIteration
        self.iterCount += 1
        return frequency_list[self.iterCount - 1]
```
